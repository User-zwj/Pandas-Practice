## Concatenate two dataframe

```python
left = pd.DataFrame({'C1': ['K0', 'K1', 'K2', 'K3'],
                     'C2': ['K0', 'K1', 'K0', 'K1'],
                      'C3': ['A0', 'A1', 'A2', 'A3'],
                      'C4': ['B0', 'B1', 'B2', 'B3']}) 

right = pd.DataFrame({'C1': ['K1', 'K2', 'K3', 'K4'],
                      'C2': ['K1', 'K1', 'K1', 'K1'],
                       'C3': ['C1', 'C2', 'C3', 'C4'],
                       'C4': ['D1', 'D2', 'D3', 'D4']})
```

```python
left
```

```python
right
```

```python
result1 = pd.concat([left,right],axis=1)
result1
```

```python
## Not recommended
result2=pd.concat([left,right], axis=0)
result2
```
